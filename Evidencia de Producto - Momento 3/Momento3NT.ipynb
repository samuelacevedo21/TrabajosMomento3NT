```python
import pandas as pd
```

```python
datos_viajes = pd.read_csv("/content/drive/MyDrive/Base de Datos.csv" , sep = ";")
```

```python
datos_viajes.head(50)
```

```python
sel_destino = datos_viajes ["Destino"] == "Amazonas"
cant_dst = datos_viajes [sel_destino].shape [0]
cant_dst
```

```python
sel_destino = datos_viajes ["Destino"] == "Amazonas"
cant_otros_dst = datos_viajes [~sel_destino].shape [0]
cant_otros_dst
```

```python
sel_cart_nuqui_amazonas = (datos_viajes ["Destino"] == " Cartagena") | (datos_viajes ["Destino"] == "Nuqui") | (datos_viajes ["Destino"] == "Amazonas")
sel_cart_nuqui_amazonas = datos_viajes[sel_cart_nuqui_amazonas].shape[0]
sel_cart_nuqui_amazonas
```

```python
sel_precio_dest = (datos_viajes ['Descuento'] == 100000) & (datos_viajes ['Descuento'] == 500000)
cant_dest_precio = datos_viajes[sel_precio_dest].shape[0]
cant_dest_precio
```

```python
sel_lugar_destino = (datos_viajes ['Destino'] == "Amazonas") & (datos_viajes ['Apellido'] == "Agudelo")
cant_lugar_destino = datos_viajes[sel_lugar_destino].shape[0]
cant_lugar_destino
```

```python
print ("Se encontraron los siguientes resultados: ")
print (f"El numero de destinos a Amazonas es de {sel_destino}")
print (f"El numero de destinos restantes es de {cant_otros_dst}")
print (f"El numero de viajes con destino Cartagena, Nuqui; AMazonas es de {sel_cart_nuqui_amazonas}")
print (f"La cantidad de destinos con descuentos iguales a $100000 y $500000 son {sel_precio_dest}")
print (f"La cantidad de destinos a AMazons con Apellidos Agudelo es de  {sel_lugar_destino}")
```
